```python
import networkx as nx
import csv 
from csv import reader
import random
import numpy as np
import matplotlib.pyplot as plt
import collections
```

```python
Data = open('ntw.csv', "r")
#ignore first line
next(Data, None)
G = nx.parse_edgelist(Data, delimiter=',',create_using=nx.DiGraph)
```

### Initial Exploration
First let's ask for some information regarding our directed graph which doesn't have any weights and let's visualize it using draw (G)

```python
nx.info(G)
```

```python
plt.figure(figsize=(20, 20))
nx.draw(G)
plt.show()
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # enumerating through the nodes to ask for degree and sorting them from biggest to smallest
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt,width=10, color='b')

plt.title("Degree Distribution")
plt.ylabel("Count")
plt.xlabel("Degree")
```

```python
degree_sequence = sorted([d for n, d in G.in_degree()], reverse=True)  # enumerating through the nodes to ask for degree and sorting them from biggest to smallest
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt,width=10, color='b')

plt.title("In-Degree Distribution")
plt.ylabel("Count")
plt.xlabel("Degree")
```
